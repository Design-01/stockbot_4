```python
import pandas as pd

def map_to_gics(input_file, column_name, mapping_type='naics', output_suffix='mapped'):
    """
    Maps industry classifications to GICS categories and corresponding ETF tickers.
    
    Parameters:
    input_file (str): Path to the input CSV/Excel file
    column_name (str): Name of the column containing industry classifications
    mapping_type (str): Type of mapping to use ('naics', 'trbc', or 'sic')
    output_suffix (str): Suffix to append to output filename
    
    Returns:
    pd.DataFrame: The mapped DataFrame
    """
    # ETF mapping dictionary
    gics_to_etf = {
        'Energy': 'XLE',
        'Materials': 'XLB',
        'Industrials': 'XLI',
        'Consumer Discretionary': 'XLY',
        'Consumer Staples': 'XLP',
        'Health Care': 'XLV',
        'Financials': 'XLF',
        'Information Technology': 'XLK',
        'Communication Services': 'XLC',
        'Utilities': 'XLU',
        'Real Estate': 'XLRE'
    }
    
    # Original classification mappings
    naics_to_gics = {
        'Agriculture, Forestry, Fishing and Hunting': ['Consumer Staples'],
        'Mining, Quarrying, and Oil and Gas Extraction': ['Energy', 'Materials'],
        'Utilities': ['Utilities'],
        'Construction': ['Industrials'],
        'Wholesale rade': ['Consumer Discretionary', 'Consumer Staples'],
        'Information': ['Information Technology', 'Communication Services'],
        'Finance and Insurance': ['Financials'],
        'Real Estate and Rental and Leasing': ['Real Estate'],
        'Professional, Scientific, and echnical Services': ['Information Technology', 'Industrials'],
        'Management of Companies and Enterprises': ['Financials'],
        'Administrative and Support and Waste Management and Remediation Services': ['Industrials'],
        'Educational Services': ['Consumer Discretionary'],
        'Health Care and Social Assistance': ['Health Care'],
        'Arts, Entertainment, and Recreation': ['Consumer Discretionary'],
        'Accommodation and Food Services': ['Consumer Discretionary'],
        'Other Services (except Public Administration)': ['Consumer Discretionary'],
        'Public Administration': ['Industrials']
    }

    trbc_to_gics = {
        'Energy': 'Energy',
        'Basic Materials': 'Materials',
        'Industrials': 'Industrials',
        'Consumer Cyclicals': 'Consumer Discretionary',
        'Consumer Non-Cyclicals': 'Consumer Staples',
        'Healthcare': 'Health Care',
        'Financials': 'Financials',
        'Technology': 'Information Technology',
        'Utilities': 'Utilities',
        'Real Estate': 'Real Estate',
        'Institutions, Associations & Organizations': 'Industrials',
        'Government Activity': 'Industrials',
        'Academic & Educational Services': 'Consumer Discretionary'
    }

    sic_to_gics = {
        'Agriculture, Forestry and Fishing': ['Consumer Staples'],
        'Mining and Quarrying': ['Energy', 'Materials'],
        'Manufacturing': ['Industrials', 'Materials', 'Consumer Staples', 'Consumer Discretionary'],
        'Electricity, gas, steam and air conditioning supply': ['Utilities'],
        'Water supply, sewerage, waste management and remediation activities': ['Utilities'],
        'Construction': ['Industrials'],
        'Wholesale and retail trade; repair of motor vehicles and motorcycles': ['Consumer Discretionary', 'Consumer Staples'],
        'Transportation and storage': ['Industrials'],
        'Accommodation and food service activities': ['Consumer Discretionary'],
        'Information and communication': ['Information Technology', 'Communication Services'],
        'Financial and insurance activities': ['Financials'],
        'Real estate activities': ['Real Estate'],
        'Professional, scientific and technical activities': ['Information Technology', 'Industrials'],
        'Administrative and support service activities': ['Industrials'],
        'Public administration and defence; compulsory social security': ['Industrials'],
        'Education': ['Consumer Discretionary'],
        'Human health and social work activities': ['Health Care'],
        'Arts, entertainment and recreation': ['Consumer Discretionary'],
        'Other service activities': ['Consumer Discretionary'],
        'Activities of households as employers; undifferentiated goods- and services-producing activities of households for own use': ['Consumer Discretionary'],
        'Activities of extraterritorial organisations and bodies': ['Industrials']
    }
    
    # Select the appropriate mapping
    mapping_dict = {
        'naics': naics_to_gics,
        'trbc': trbc_to_gics,
        'sic': sic_to_gics
    }[mapping_type.lower()]
    
    try:
        # Try to read the file as CSV first
        try:
            df = pd.read_csv(input_file)
        except:
            # If CSV fails, try Excel
            df = pd.read_excel(input_file)
        
        # Create the new GICS mapping column
        if mapping_type.lower() == 'trbc':
            # TRBC mapping returns single values
            df['gics_map'] = df[column_name].map(mapping_dict)
            # Map ETF tickers directly
            df['etf_ticker'] = df['gics_map'].map(gics_to_etf)
        else:
            # NAICS and SIC mappings return lists
            df['gics_map'] = df[column_name].map(mapping_dict)
            
            # Function to get ETF tickers for multiple GICS sectors
            def get_etf_tickers(gics_list):
                if isinstance(gics_list, list):
                    return ', '.join([gics_to_etf.get(gics, '') for gics in gics_list])
                return ''
            
            # Convert GICS list to string and map ETF tickers
            df['gics_map'] = df['gics_map'].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
            df['etf_ticker'] = df['gics_map'].apply(lambda x: get_etf_tickers(x.split(', ')) if x else '')
        
        # Create output filename
        filename_parts = input_file.rsplit('.', 1)
        output_file = f"{filename_parts[0]}_{output_suffix}.{filename_parts[1]}"
        
        # Save the file in the same format as input
        if input_file.endswith('.csv'):
            df.to_csv(output_file, index=False)
        else:
            df.to_excel(output_file, index=False)
            
        print(f"Mapped file saved as: {output_file}")
        return df
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")
        return None



# For NAICS mapping:
df1 = map_to_gics('TRBC.csv',  'Sector Name', mapping_type='trbc',  output_suffix='mapped')
df2 = map_to_gics('NIACS.csv', 'Sector Name', mapping_type='naics', output_suffix='mapped')
df3 = map_to_gics('SIC.csv',   'Section Name', mapping_type='sic',  output_suffix='mapped')
display(df1)
display(df2)
display(df3)

```

```python
import pandas as pd

def add_etf_tickers_to_gics(input_file, gics_column_name, output_suffix='with_etf'):
    """
    Adds corresponding ETF tickers to a GICS classification file.
    
    Parameters:
    input_file (str): Path to the input CSV/Excel file containing GICS classifications
    gics_column_name (str): Name of the column containing GICS classifications
    output_suffix (str): Suffix to append to output filename
    
    Returns:
    pd.DataFrame: The DataFrame with added ETF tickers
    """
    # ETF mapping dictionary
    gics_to_etf = {
        'Energy': 'XLE',
        'Materials': 'XLB',
        'Industrials': 'XLI',
        'Consumer Discretionary': 'XLY',
        'Consumer Staples': 'XLP',
        'Health Care': 'XLV',
        'Financials': 'XLF',
        'Information Technology': 'XLK',
        'Communication Services': 'XLC',
        'Utilities': 'XLU',
        'Real Estate': 'XLRE'
    }
    
    try:
        # Try to read the file as CSV first
        try:
            df = pd.read_csv(input_file)
        except:
            # If CSV fails, try Excel
            df = pd.read_excel(input_file)
        
        # Add ETF ticker column
        df['etf_ticker'] = df[gics_column_name].map(gics_to_etf)
        
        # Create output filename
        filename_parts = input_file.rsplit('.', 1)
        output_file = f"{filename_parts[0]}_{output_suffix}.{filename_parts[1]}"
        
        # Save the file in the same format as input
        if input_file.endswith('.csv'):
            df.to_csv(output_file, index=False)
        else:
            df.to_excel(output_file, index=False)
            
        print(f"File with ETF tickers saved as: {output_file}")
        return df
        
    except Exception as e:
        print(f"Error processing file: {str(e)}")
        return None

# Example usage:

df = add_etf_tickers_to_gics('GICS.csv', 'Sector Name', output_suffix='mapped')
df
```

```python
import pandas as pd

def get_etf_from_sector_code(classification, sector_code):
    file_names = {
        'NIACS': 'NIACS_mapped.csv',
        'TRBC': 'TRBC_mapped.csv',
        'SIC_US': 'SIC_US_mapped.csv',
        'GICS' : 'GICS_mapped.csv'
    }

    code_column = {
        'NIACS': 'Sub-Industry Code',
        'TRBC': 'Hierarchical ID',
        'SIC_US': 'SIC',
        'GICS': 'Sub-Industry Code'
    }

    sector_column = {
        'NIACS': 'Sub-Industry Name',
        'TRBC': 'Sub-Industry Name',
        'SIC_US': 'Description',
        'GICS': 'Sub-Industry Name'
    }

    df = pd.read_csv(file_names[classification])
    etf    = df[df[code_column[classification]]   == sector_code]['etf_ticker'].values[0]
    sector = df[df[code_column[classification]] == sector_code][sector_column[classification]].values[0]
    print(f"ETF for {classification} sector {sector} ({sector_code}): {etf}")
    return etf

# Example usage:

print(get_etf_from_sector_code('TRBC', 5310101014)) 
print(get_etf_from_sector_code('NIACS', 335312)) 
print(get_etf_from_sector_code('SIC_US', 3711)) 

```

```python
import pandas as pd

"""https://github.com/saintsjd/sic4-list/blob/master/sic-codes.csv

this contains all the groups and subgroups of the SIC codes and descriotion of each group. 
Look in the left of the githuib page to download the csv files for each set of codes.
"""

# Read all CSV files
div = pd.read_csv('divisions.csv')
mg = pd.read_csv('major-groups.csv')
ig = pd.read_csv('industry-groups.csv')
sic = pd.read_csv('sic-codes.csv')

# Rename columns to avoid conflicts and make merging easier
div = div.rename(columns={
    'Description': 'Division Description',
    'Full Description': 'Division Full Description'
})

mg = mg.rename(columns={
    'Description': 'Major Group Description'
})

ig = ig.rename(columns={
    'Description': 'Industry Group Description'
})

# Perform the merges
# First merge SIC with Industry Groups
merged_df = pd.merge(
    sic,
    ig[['Division', 'Major Group', 'Industry Group', 'Industry Group Description']],
    how='left',
    on=['Division', 'Major Group', 'Industry Group']
)

# Then merge with Major Groups
merged_df = pd.merge(
    merged_df,
    mg[['Division', 'Major Group', 'Major Group Description']],
    how='left',
    on=['Division', 'Major Group']
)

# Finally merge with Divisions
merged_df = pd.merge(
    merged_df,
    div[['Division', 'Division Description', 'Division Full Description']],
    how='left',
    on='Division'
)

# Reorder columns according to your new_columns list
new_columns = [
    'Division', 'Division Description', 
    'Major Group', 'Major Group Description',
    'Industry Group', 'Industry Group Description',
    'SIC', 'Description'
]

final_df = merged_df[new_columns]

# Save the result to a new CSV file
final_df.to_csv('SIC_US.csv', index=False)

# Display the first few rows to verify the result
print("Enhanced SIC Codes Preview:")
display(final_df.head())
final_df['Division Description'].unique()
```
