```python
def insert_tracker_data(df, whereColumn:str, whereValue:str, insertAtColumn:float, insertValue:float):
    df.loc[df[whereColumn] == whereValue, insertAtColumn] = insertValue
    return df

import pandas as pd
df = pd.DataFrame({
    'symbol': ['TSLA', 'ROKU', 'SOFI'],
    'allowedETF': [True, True, False]
})

insert_tracker_data(df, 'symbol', 'TSLA', 'bob', False)
insert_tracker_data(df, 'symbol', 'ROKU', 'bob', 20)
```
