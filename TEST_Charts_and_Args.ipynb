```python
#  Touch Scores

from dataclasses import dataclass, field
import math
import pandas as pd




@dataclass
class TouchScore:
    name: str = 'TouchScore' # assign score to this colname 
    levelTol: float = 0.25 #  multiple of the ATR so 0.5 = 25% of ATR above and 25% below the key level
    timeDecay: float = 0.001 #  decay factor for time, 0.01 = 1% decay per bar
    volChgCol: str = 'volChg' #  column name for volume change
    atrCol: str = 'atr' #  column name for ATR
    weights:       dict = field(default_factory=lambda: {'tail': 0.35, 'vol': 0.25, 'rejection': 0.2, 'span': 0.1, 'decay': 0.1, 'confluence': 0.2})
    confBonThresh: dict = field(default_factory=lambda: {'tail': 0.35, 'vol': 0.25, 'rejection': 0.2, 'span': 0.2})

    def __post_init__(self):
        self.bucket_list = [] # list of buckets to store the data   
        self.name_tails = f'{self.name}_tails'
        self.name_vols = f'{self.name}_vols'
        self.name_rejections = f'{self.name}_rejections'
        self.name_spans = f'{self.name}_spans'
        self.name_confluences = f'{self.name}_confluences'  


    def get_reversed_records(self, df):
        #  reverse the records so we can work backwards
        #  and also to get the correct order of the records
        dfx = df.copy()
        records = dfx.reset_index().to_dict('records') #  for debugging to include the index
        records.reverse()
        return records
    
    def get_touch_zone(self, price, atr):
        zone_val = self.levelTol * atr / 2
        top_edge = price + zone_val
        bot_edge = price - zone_val
        return round(top_edge,2), round(bot_edge,2)

    def get_bounce_direction(self, i):
        #  get the last bucket
        pos_before = self.bucket_list[i-1]['pos']
        pos_after = self.bucket_list[i+1]['pos']
        if pos_before == 'above' and pos_after == 'above':
            return 'up'
        if pos_before == 'below' and pos_after == 'below':
            return 'down'
        return 'transition'
    
    def get_rejection_strenght(self, i):
        max_before = self.bucket_list[i-1]['max']
        max_after = self.bucket_list[i+1]['max']
        return (max_before + max_after) / 2
    
    def normalise(self, value, min_val, max_val):
        if min_val == max_val:
            return 0
        return min(1, (value - min_val) / (max_val - min_val))

    
    def compute_metrics(self, df):
        if len(self.bucket_list) < 3:
            return
        
        for i, b in enumerate(self.bucket_list[:-1]):
            if b['stage'] == 'Touch':
                b['bounceDir'] = self.get_bounce_direction(i)
                if b['bounceDir'] == 'transition':
                    continue

                b['rejection'] = self.get_rejection_strenght(i)
                
                tailLen   = b['toptailmax'] if b['pos'] == 'below' else b['bottailmax'] 
     
                b['norm_tailLen']   = self.normalise(tailLen, 0, df['ATR_14'].iat[-1]) * self.weights['tail']
                b['norm_vol']       = self.normalise(b['vol'] , 0, df['volume'].mean()) * self.weights['vol']
                b['norm_rejection'] = self.normalise(b['rejection'] , 0, 5) * self.weights['rejection']
                b['norm_span']      = self.normalise(b['span'] , 0, 5) * self.weights['span']
                b['norm_decay']     = round((self.weights['decay'] - self.normalise(b['decay'], 0, len(df))) * self.weights['decay'], 5)

                b['score'] = (b['norm_tailLen'] + b['norm_vol'] + b['norm_rejection'] + b['norm_span'] + b['norm_decay']) 

                confluence_count = 0
                for k, v in self.confBonThresh.items():
                    if b[k] > v:
                        confluence_count += 1 

                # Calculate percentage of thresholds met and apply proportional bonus
                confluence_ratio = confluence_count / len(self.confBonThresh)
                b['scoreWconf'] = b['score'] + confluence_ratio * self.weights['confluence']

    def get_column_means(self, df, columns):
        """
        Calculate the mean of specified columns in a DataFrame.
        
        Parameters:
        df (pandas.DataFrame): The DataFrame containing the data
        columns (list): List of column names to calculate means for
        
        Returns:
        dict: Dictionary with column names as keys and their mean values as values
        """
        means = {}
        
        for col in columns:
            if col in df.columns:
                means[col] = round(df[col].mean(),2)
            else:
                means[col] = None  # or you could raise an exception
                
        return means
    
    def get_last_df_row(self):
        return pd.DataFrame(self.bucket_list)
    
    def run(self, df, keyLevel=None):   

        records = self.get_reversed_records(df)
        stage = 'starting'
        atr = (df.high-df.low).mean()

        for count, row in enumerate(records):
            # Get values using dictionary access
            idx = row['date']
            # lev = row['Sup_1']
            lev = keyLevel
            op = row['open']
            cl = row['close']
            hi = row['high']
            lo = row['low']
            vol = row['volume']
            top_edge, bot_edge = self.get_touch_zone(lev, atr)

            if count == 0:
                #  first row, so set the top and bottom edges
                top_edge, bot_edge = self.get_touch_zone(lev, atr)
                print(f"Row: {count}, {idx} : {top_edge=}, {bot_edge=}")

            #  claude says this the fastest way to check for NaN but must be a float
            if isinstance(lev, float) and math.isnan(lev):
                continue
            
            touch_from_above = cl > top_edge and lo <= top_edge
            touch_from_below = cl < bot_edge and hi >= bot_edge
            is_above = lo > top_edge
            is_below = hi < bot_edge
            in_zone = hi >= bot_edge and lo <= top_edge

            # if in_zone:
            #     print(f"Row: {count}, {idx} : Stage: {stage}, Touch from above: {touch_from_above}, Touch from below: {touch_from_below}, In zone: {in_zone} {lev=}, {round(top_edge,4)=}, {round(bot_edge,4)=}")


            # === Detremine trasitions to stages first ===

            # start filling between
            if not in_zone and stage == 'starting':
                # bucket filling
                stage = 'Between'
                self.bucket_list += [{'bounceDir':'', 'start': idx, 'end': None, 'stage': stage, 'pos': None, 'span': 0, 'max': 0}]

            # end fillig Between - start filling touches
            if in_zone and stage == 'Between':
                stage = 'Touch'
                self.bucket_list += [{'bounceDir': '', 'start': idx, 'end': None, 'stage': stage, 'decay': 0, 'vol': vol, 'span': 0, 'tail': 0}]
            
            # end filling touches - start filling Between
            if not in_zone and stage == 'Touch':
                stage = 'Between'
                self.bucket_list += [{'bounceDir':'', 'start': idx, 'end': None, 'stage': stage, 'pos': None, 'span': 0, 'max': 0}]
            
            # ==== fiil once decided on the stage ====
            
            # continue filling Between
            if not in_zone and stage == 'Between':
                b = self.bucket_list[-1]
                b['span'] +=1
                b['end'] = idx


                if is_above:
                    # touch from above
                    b['pos'] = 'above'
                    b['max'] = max(b['max'], lo - top_edge)


                elif is_below:
                    # touch from below
                    b['pos'] = 'below'
                    b['max'] = max(b['max'], bot_edge - hi)


            # continue filling touches
            if in_zone and stage == 'Touch':
                b = self.bucket_list[-1]
                b['end'] = idx
                b['span'] += 1
                

                if touch_from_above and not is_above:
                    # touch from above
                    b['pos'] = 'above'
                    b['decay'] = self.timeDecay * count
                    b['vol'] += vol
                    b['bottail'] = min(op, cl) - lo
                    b['bottailmax'] = max(b['bottailmax'], b['bottail']) if 'bottailmax' in b else b['bottail']

                elif touch_from_below and not is_below:
                    # touch from below
                    b['pos'] = 'below'
                    b['decay'] = self.timeDecay * count
                    b['vol'] += vol
                    b['toptail'] = hi - max(op, cl)
                    b['toptailmax'] = max(b['toptailmax'], b['toptail']) if 'toptailmax' in b else b['toptail']

                # if b['span'] > 1:
                #     print(f"Row: {count}, {idx} : {in_zone=} {is_above=}, {is_below=} , {top_edge=}, {bot_edge=}, {b}")
             


        return self.bucket_list




            
pd.set_option('display.max_columns', None)
ts = TouchScore(levelTol=0.5)
ts.run(f.data, keyLevel=537)
ts.compute_metrics(f.data)
dfscores = ts.get_last_df_row()
ts.get_column_means(dfscores, ['score', 'scoreWconf', 'norm_tailLen', 'norm_vol', 'norm_rejection', 'norm_span', 'norm_decay'])
```

```python
import pandas as pd
import numpy as np
import time

# Create a sample DataFrame with similar structure to your data
# Using a larger size to make timing differences more apparent
size = 10_000
df = pd.DataFrame({
    'time': pd.date_range(start='2023-01-01', periods=size, freq='1min'),
    'close': np.random.random(size) * 100,
    'key_level': np.random.random(size) * 100,
    'atr': np.random.random(size) * 5,
    'volume': np.random.randint(1000, 10000, size)
})

print(f"DataFrame size: {len(df)} rows x {len(df.columns)} columns")

# Method 1: Using iloc in a loop (original approach)
def method1_iloc_in_loop():
    start_time = time.time()
    dummy_sum = 0
    
    for idx in range(len(df) - 1, -1, -1):
        row = df.iloc[idx]
        # Just do something minimal with the row to ensure it's accessed
        dummy_sum += row['close']
    
    end_time = time.time()
    return end_time - start_time, dummy_sum

# Method 2: Using itertuples() with reversed DataFrame
def method2_itertuples():
    start_time = time.time()
    dummy_sum = 0
    
    reversed_df = df.iloc[::-1].reset_index(drop=True)
    for row in reversed_df.itertuples():
        # Access via attribute
        dummy_sum += row.close
    
    end_time = time.time()
    return end_time - start_time, dummy_sum

# Method 3: Convert to records and reverse
def method3_to_dict_records():
    start_time = time.time()
    dummy_sum = 0
    
    records = df.to_dict('records')
    records.reverse()
    for row in records:
        # Access via dictionary key
        dummy_sum += row['close']
    
    end_time = time.time()
    return end_time - start_time, dummy_sum

# Run the benchmarks
print("\nRunning benchmarks (this may take a moment)...")

print("\nMethod 1: iloc in loop")
time1, sum1 = method1_iloc_in_loop()
print(f"Time: {time1:.4f} seconds")
print(f"Verification sum: {sum1:.2f}")

print("\nMethod 2: itertuples with reversed DataFrame")
time2, sum2 = method2_itertuples()
print(f"Time: {time2:.4f} seconds")
print(f"Verification sum: {sum2:.2f}")
print(f"Speedup vs Method 1: {time1/time2:.2f}x faster")

print("\nMethod 3: to_dict('records') with reverse")
time3, sum3 = method3_to_dict_records()
print(f"Time: {time3:.4f} seconds")
print(f"Verification sum: {sum3:.2f}")
print(f"Speedup vs Method 1: {time1/time3:.2f}x faster")

print("\nSummary:")
fastest = min([(time1, "Method 1 (iloc)"), (time2, "Method 2 (itertuples)"), (time3, "Method 3 (records)")])
print(f"Fastest method: {fastest[1]} - {fastest[0]:.4f} seconds")

# For smaller DataFrames, let's check if the results differ
small_df = df.iloc[:10000].copy()
print("\n\nTesting with smaller DataFrame (10,000 rows)...")

# Method 1 with small DataFrame
start_time = time.time()
for idx in range(len(small_df) - 1, -1, -1):
    row = small_df.iloc[idx]
    # Just do something with the row
    _ = row['close']
small_time1 = time.time() - start_time

# Method 2 with small DataFrame
start_time = time.time()
reversed_small_df = small_df.iloc[::-1].reset_index(drop=True)
for row in reversed_small_df.itertuples():
    # Access via attribute
    _ = row.close
small_time2 = time.time() - start_time

# Method 3 with small DataFrame
start_time = time.time()
small_records = small_df.to_dict('records')
small_records.reverse()
for row in small_records:
    # Access via dictionary key
    _ = row['close']
small_time3 = time.time() - start_time

print(f"Small DataFrame - Method 1 (iloc): {small_time1:.4f} seconds")
print(f"Small DataFrame - Method 2 (itertuples): {small_time2:.4f} seconds")
print(f"Small DataFrame - Method 3 (records): {small_time3:.4f} seconds")

small_fastest = min([(small_time1, "Method 1 (iloc)"), (small_time2, "Method 2 (itertuples)"), (small_time3, "Method 3 (records)")])
print(f"Fastest method for small DataFrame: {small_fastest[1]} - {small_fastest[0]:.4f} seconds")
```

```python
def compute_weighted_score(metrics, weights):
    # metrics should be a dictionary with keys for each metric
    # weights should be a dictionary with the same keys, representing percentages
    
    # Ensure weights sum to 100%
    total_weight = sum(weights.values())
    if abs(total_weight - 100) > 0.01:
        print(f"Warning: Weights sum to {total_weight}%, not 100%")
    
    # Compute weighted score
    weighted_score = 0
    for metric, value in metrics.items():
        if metric in weights:
            weighted_score += value * (weights[metric] / 100)
    
    return weighted_score

# Example usage:
metrics = {
    "volume": 2500,
    "tail_size": 0.85,
    "rejection_size": 1.2,
    "span": 45,
    "time_decay": 0.65
}

weights = {
    "volume": 35,
    "tail_size": 25,
    "rejection_size": 15,
    "span": 10,
    "time_decay": 15
}

score = compute_weighted_score(metrics, weights)
print(f"Weighted score: {score}")
```
